```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Loading and storing the csv
weather_data_to_load = "../WeatherPy/weather_data.csv"
weather_data = pd.read_csv(weather_data_to_load)
weather_data
```

```python
#Cleaning weather data by removing data with humidity more than 100
index_names = weather_data[ weather_data['Humidity'] > 100 ].index
clean_weather_data = weather_data.drop(index_names, inplace = False)
clean_weather_data
```

```python
#drop na values
clean_weather_data.dropna(inplace = True) 
clean_weather_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

```python
# Store latitude and longitude in locations
locations = clean_weather_data[["Latitude", "Longitude"]]
locations
```

```python
# Fill NaN values and convert to float
humidity = clean_weather_data["Humidity"].astype(float)
humidity
```

```python
# Plot Heatmap
fig=gmaps.figure(center=(20.0, 10.0),zoom_level=1.7)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2.5)


# Add layer
fig.add_layer(heat_layer)


# Display figure
fig

```

### Create new DataFrame fitting weather criteria

```python
#Filtering data with temperature > 21 and < 27
index_names = clean_weather_data[ (clean_weather_data['Temp'] > 21) & (clean_weather_data['Temp'] < 27)].index
fit_weather_data = clean_weather_data.drop(index_names, inplace = False)
fit_weather_data
```

```python
#Filtering data with wind speed less than 4.5 m/s
index_names = fit_weather_data[ fit_weather_data['Wind Speed'] <= 4.5].index
fit_weather_data = fit_weather_data.drop(index_names, inplace = False)
fit_weather_data
```

```python
#Filtering data with zero cloudiness
index_names = fit_weather_data[ fit_weather_data['Cloudiness'] > 0].index
fit_weather_data = fit_weather_data.drop(index_names, inplace = False)
fit_weather_data
```

### Hotel Map

```python
hotel_df = fit_weather_data[['City', 'Country', 'Latitude', 'Longitude', 'Humidity']]
hotel_df['Hotel Name'] = ''
hotel_df
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius": 10000,
    "keyword": "hotel",
    "type": "hotel",
    "key": g_key
    }
# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():

    # get city from df
    city = row['City']
    
    params['location'] = f'{row["Latitude"]},{row["Longitude"]}'

    # assemble url and make API request
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("na")
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
marker_layer=gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(marker_layer)
# Display figure
fig
```
